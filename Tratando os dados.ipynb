```python
def take_ticket(list_of_tickets): 
    # take the tickets in the beggining of the name
    name = []
    
    for char in list_of_tickets:
        if char != "_":
            name.append(char)
        else:
            break
            
    ticket = ''.join([str(elem) for elem in name])
    
    return ticket
```

```python
list_of_tickets = []
growth_quartely = pd.DataFrame()
for file in files:
    if file.endswith('_Growth_Quartely.xlsx'):
        ticket = take_ticket(file)
        print(file)
        table = pd.read_excel('Stock Rows Data/{}'.format(file))
        table['Ticket'] = ticket
        table.rename(columns = {'Unnamed: 0': 'Indicators'}, inplace = True)
        growth_quartely = growth_quartely.append(table, ignore_index=True)


growth_quartely.head()       
```
